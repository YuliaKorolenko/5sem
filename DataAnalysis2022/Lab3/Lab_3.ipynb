```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from math import sqrt
```

```python
df = pd.read_csv("andan-2022-autumn-lab-3/train.csv")
df_test = pd.read_csv("andan-2022-autumn-lab-3/test_x.csv")
df.head(100)
```

```python
cor = df.corr().abs()
cor = cor.get("rating_org").sort_values()
```

```python
print(cor[cor > 0.09].sort_values(ascending=False))
```

```python
df["comfort_dis_my"]=df["wheelchair_access"]+df["parking_disabled"]+df["toilet_for_disabled"]+df["elevator_wheelchair_accessible"]+df["wheelchair_accessible"]
df["comfort_my"]=df["automatic_door"]+df["elevator"]+df["call_button"]+df["air_conditioning"]+df["car_park"]
df["internet_my"]=max(df["wi_fi"]+df["internet access"])
df["takeaway_my"]=df["online_takeaway"]+df["takeaway"]
df["card_my"]=max(df["payment_by_credit_card"]+df["accepted_credit_cards"])
df["delivery_my"]=df["food_product_delivery"]+df["food_delivery"]+df["delivery"]+df["free_delivery"]
```

```python
df_test["comfort_dis_my"]=df_test["wheelchair_access"]+df_test["parking_disabled"]+df_test["toilet_for_disabled"]+df_test["elevator_wheelchair_accessible"]+df_test["wheelchair_accessible"]
df_test["comfort_my"]=df_test["automatic_door"]+df_test["elevator"]+df_test["call_button"]+df_test["air_conditioning"]+df_test["car_park"]
df_test["internet_my"]=max(df_test["wi_fi"]+df_test["internet access"])
df_test["takeaway_my"]=df_test["online_takeaway"]+df_test["takeaway"]
df_test["card_my"]=max(df_test["payment_by_credit_card"]+df_test["accepted_credit_cards"])
df_test["delivery_my"]=df_test["food_product_delivery"]+df_test["food_delivery"]+df_test["delivery"]+df_test["free_delivery"]
```

```python
X = pd.DataFrame(df, columns=["user_id","org_id","rating","ts","food_delivery","breakfast","takeaway","summer_terrace","business_lunch","payment_by_credit_card","coffee_to_go","closed_for_quarantine","online_takeaway","karaoke","special_menu","sports_broadcasts","food_court1","craft_beer","projector","gift_certificate","automatic_door","food_product_delivery","car_park","promotions","cafe","minimum_order","free_delivery","self_service_kiosks","has_restaurant","air_conditioning","pets","elevator","has_bar","privilege_tickets","organic_food","vip_zone","hall_capacity","face_control","show_program","music","chillout","dancefloor","dress_code","billiards","strip","ramp","around_the_clock_work1","pickup","farm_products","delivery","coffee_shop","call_button","table_games","vinotheque","cakes_for_different_occasions","accepted_credit_cards","kalyan","teahouse","bread_from_tandoor","handmade_goods"])
X_test = pd.DataFrame(df_test, columns=["user_id","org_id","rating","ts","food_delivery","breakfast","takeaway","summer_terrace","business_lunch","payment_by_credit_card","coffee_to_go","closed_for_quarantine","online_takeaway","karaoke","special_menu","sports_broadcasts","food_court1","craft_beer","projector","gift_certificate","automatic_door","food_product_delivery","car_park","promotions","cafe","minimum_order","free_delivery","self_service_kiosks","has_restaurant","air_conditioning","pets","elevator","has_bar","privilege_tickets","organic_food","vip_zone","hall_capacity","face_control","show_program","music","chillout","dancefloor","dress_code","billiards","strip","ramp","around_the_clock_work1","pickup","farm_products","delivery","coffee_shop","call_button","table_games","vinotheque","cakes_for_different_occasions","accepted_credit_cards","kalyan","teahouse","bread_from_tandoor","handmade_goods"])
Y = np.array(df['rating_org'])
```

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=5000)
```

```python
model.fit(X, Y)
```

```python
pred_y = model.predict(X_test)
```

```python
print(pred_y)
```

```python
answer = pd.DataFrame({'rating_org' : pred_y})
answer.index.name = 'id'
print(answer)
```
